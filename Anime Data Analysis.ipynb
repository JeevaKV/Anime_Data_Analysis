## **Investigating the Non Numerical data**
Since the numerical data columns doesn't give us any relation,therefore we are going to use the non numeric data columns.

```python
genre=new_df[' genre'].unique()
genre
```

Let's see no of columns unique in the genre.
**Note**:The genre is tuple which consists of datatypes of objects.

```python
genre=np.array(new_df[' genre'].unique())
genre.shape
```

```python
for g in genre:
    print(g)
    print(len(g))
```

```python
import ast
gd = {}
ind=0;
for genres in genre:
    genreList = ast.literal_eval(genres) 
    print("Length ",ind,": ", len(genreList))
    for i in genreList:
        if i in gd:     
            gd[i]+=1
        else:
            gd[i]=1       
        print(i)
    print(" ")
    ind+=1
```

```python
gd.keys()
```

Now lets calculate the total no of genres present in the given dataset

```python
print("No of genres in the given data set:",len(gd))
```

Let's sort the values in the dictionary for the purpose of easy computation and plotting in the 2D-plane

```python
gd_sorted={keys: values for keys, values in sorted(gd.items(), key=lambda item: item[1])}
gd_sorted
```

```python
pk,pv=[],[]
for k in gd:
    print("k: %s , v: %s" % (k, gd[k]))
    pk.append(k)
    pv.append(gd[k])
pk,pv
```

Let's use a histogram to represent the data of genre we have been working through in this section.

```python
f = plt.figure(figsize=(20,10))
x = f.add_axes([0,0,1,1])
y = np.arange(len(pk))
x.barh(y,pv)
x.set_yticks(y)
x.set_yticklabels(pk) 
x.set_xlabel('Count')
x.set_title('Genre')

plt.show()
```

From the histogram,we can conclude that the most viewed genre anime is **Comdey**. And the least viewed anime is of genre **Cars**

```python
new_df[' producers'].unique()
```

```python
producer=np.array(new_df[' producers'].unique())
```

```python
pd = {}

index=0;
for i in producer:
    producerList = ast.literal_eval(i) 
    print("Length ",index,": ", len(producerList))
    for j in producerList:
        if j in pd:      
            pd[j]+=1
        else:
            pd[j]=1       
        print(producer)
    print("\n")
    index+=1
```

```python
pd={k: v for k, v in sorted(pd.items(), key=lambda item: item[1])}
pd
```

```python
List=[]
for key, value in pd.items():
    t = [key,value]
    List.append(t)
print(List)
```

```python
import pandas as p
producer_df=p.DataFrame(List, columns = ["Producer", "Count"]) 
```
